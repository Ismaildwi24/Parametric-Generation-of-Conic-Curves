# Program Kurva Parabola

# Parametric Generation of Conic Curves

# Deskripsi
Program ini digunakan untuk membangkitkan kurva parabola menggunakan persamaan parametrik serta membandingkan pengaruh nilai Δt (delta t) terhadap jumlah titik dan kualitas visualisasi kurva.

# Tujuan
1. Membuat kurva parabola menggunakan persamaan parametrik.
2. Menampilkan titik-titik pembentuk kurva.
3. Membandingkan resolusi kurva berdasarkan nilai Δt.
4. Menunjukkan bahwa semakin kecil Δt maka semakin banyak titik yang dihasilkan sehingga kurva menjadi lebih halus (smooth).

# Dasar Teori
Parabola merupakan salah satu kurva irisan kerucut (Conic Section) yang memiliki eksentrisitas e = 1. Pada program ini parabola dibangkitkan menggunakan persamaan parametrik:

x = xp + a·t²

y = yp + 2·a·t

Keterangan:
- xp = koordinat x titik puncak (vertex)
- yp = koordinat y titik puncak (vertex)
- a = konstanta pembentuk parabola
- t = parameter pembentuk kurva
- Δt = interval perubahan parameter t

# Pengaruh Delta t terhadap Resolusi
Parameter t dibangkitkan dari interval -10 sampai 10. Nilai Δt menentukan jarak perubahan parameter t.

Contoh:
- Δt = 1 menghasilkan 21 titik.
- Δt = 0.5 menghasilkan 41 titik.

Semakin kecil nilai Δt maka jumlah titik yang dihasilkan semakin banyak sehingga kurva menjadi lebih halus (smooth).

# Algoritma Program
1. Menginput nilai xp, yp, dan a.
2. Menginput nilai Δt besar dan Δt kecil.
3. Membentuk parameter t dari -10 hingga 10.
4. Menghitung koordinat x dan y menggunakan persamaan parametrik.
5. Menampilkan tabel perhitungan titik-titik kurva.
6. Menggambar kurva parabola beserta titik-titiknya.
7. Membandingkan hasil resolusi rendah dan resolusi tinggi.

# Kesimpulan
Kurva parabola dapat dibangkitkan menggunakan persamaan parametrik dengan memanfaatkan parameter t. Nilai Δt memengaruhi jumlah titik yang dihasilkan. Semakin kecil nilai Δt maka semakin banyak titik yang terbentuk sehingga visualisasi kurva menjadi lebih halus tanpa mengubah bentuk dasar parabola.

```python
"""
====================================================================
IMPLEMENTASI REPRESENTASI PARAMETRIK KURVA PARABOLA
====================================================================
Mata Kuliah : Grafika Komputer - Semester 4
Topik       : Pembangkitan Kurva Konik Parametrik
Kurva       : Parabola (Eksentrisitas e = 1)

Persamaan Parametrik:
    x(t) = xp + a * t**2
    y(t) = yp + 2 * a * t

Keterangan:
    (xp, yp) = Titik puncak / vertex parabola
    a        = Parameter fokus
    t        = Parameter bebas
====================================================================
"""

import numpy as np
import matplotlib.pyplot as plt
import sys

plt.style.use('dark_background')

# ====================================================================
# BAGIAN 1 — INPUT PARAMETER
# ====================================================================

def input_parameter():
    print()
    print("=" * 68)
    print("  GENERATOR KURVA PARABOLA PARAMETRIK")
    print("  Grafika Komputer — Semester 4")
    print("=" * 68)
    print()

    xp        = float(input("  Titik puncak X  (xp)          : "))
    yp        = float(input("  Titik puncak Y  (yp)          : "))
    a         = float(input("  Parameter fokus (a)            : "))
    print()
    
    t_awal    = float(input("  t awal                         : "))
    t_akhir   = float(input("  t akhir                        : "))
    print()
    
    step_rendah = float(input("  Delta t resolusi rendah (besar) : "))
    step_tinggi = float(input("  Delta t resolusi tinggi (kecil) : "))
    print()

    return {
        "xp"          : xp,
        "yp"          : yp,
        "a"           : a,
        "t_awal"      : t_awal,
        "t_akhir"     : t_akhir,
        "step_rendah" : step_rendah,
        "step_tinggi" : step_tinggi,
    }

# ====================================================================
# BAGIAN 2 — VALIDASI INPUT
# ====================================================================

def validasi_input(params):
    a           = params["a"]
    t_awal      = params["t_awal"]
    t_akhir     = params["t_akhir"]
    step_rendah = params["step_rendah"]
    step_tinggi = params["step_tinggi"]

    errors = []
    if a == 0:
        errors.append("  [ERROR] Parameter a tidak boleh nol.")
    if t_awal >= t_akhir:
        errors.append("  [ERROR] t awal harus lebih kecil dari t akhir.")
    if step_rendah <= 0:
        errors.append("  [ERROR] Delta t resolusi rendah harus positif.")
    if step_tinggi <= 0:
        errors.append("  [ERROR] Delta t resolusi tinggi harus positif.")
    if step_tinggi >= step_rendah:
        errors.append("  [ERROR] Delta t tinggi harus < Delta t rendah.")

    if errors:
        print("\n" + "=" * 68 + "\n  VALIDASI GAGAL\n" + "=" * 68)
        for e in errors:
            print(e)
        print()
        sys.exit(1)

# ====================================================================
# BAGIAN 3 — TAMPILKAN INFORMASI PARAMETER
# ====================================================================

def tampilkan_informasi(params, t, label_resolusi):

    xp = params["xp"]
    yp = params["yp"]
    a = params["a"]

    t_awal = params["t_awal"]
    t_akhir = params["t_akhir"]

    step = t[1] - t[0] if len(t) > 1 else 0

    focus_x = xp + a
    focus_y = yp

    arah_bukaan = "Kanan (a > 0)" if a > 0 else "Kiri (a < 0)"

    jumlah_titik = len(t)

    print("=" * 90)
    print(f"INFORMASI PARAMETER PARABOLA - {label_resolusi}")
    print("=" * 90)

    print(f"""
VERTEX
------
({xp}, {yp})

FOCUS
-----
Focus = (xp + a, yp)

Focus = ({xp} + {a}, {yp})

Focus = ({focus_x}, {focus_y})

PARAMETER FOKUS
---------------
a = {a}

PERSAMAAN PARAMETRIK
--------------------
x(t) = {xp} + {a}·t²
y(t) = {yp} + 2·{a}·t

RENTANG PARAMETER
-----------------
t awal  = {t_awal}
t akhir = {t_akhir}
Delta t = {step}

JUMLAH TITIK
------------
((t akhir - t awal) / Delta t) + 1

= (({t_akhir} - ({t_awal})) / {step}) + 1

= {jumlah_titik} titik

ITERASI PARAMETER t
-------------------
Rentang t = [{t_awal}, {t_akhir}]

Parameter t =

{t}

Jumlah Iterasi = {jumlah_titik}

ARAH BUKAAN
-----------
{arah_bukaan}
""")

# ====================================================================
# BAGIAN 4 — PARAMETER & KOORDINAT
# ====================================================================

def buat_parameter_t(t_awal, t_akhir, step):
    t = np.arange(t_awal, t_akhir + step, step)
    return t[t <= t_akhir + 1e-9]

def hitung_koordinat(params, t):
    xp, yp, a = params["xp"], params["yp"], params["a"]
    x = xp + a * (t ** 2)
    y = yp + 2 * a * t
    return x, y


# ====================================================================
# BAGIAN 5 — PROSES ITERASI
# ====================================================================

def tampilkan_iterasi(params, t, x, y):

    print("\n" + "=" * 90)
    print("PROSES ITERASI PARAMETER t")
    print("=" * 90)

    step = t[1] - t[0] if len(t) > 1 else 0

    print(f"""
Rentang t : [{params['t_awal']}, {params['t_akhir']}]

Delta t   : {step}

Parameter t :

{t}

Jumlah Iterasi = {len(t)}

Karena setiap nilai t menghasilkan
1 titik koordinat (x,y),

maka jumlah titik = jumlah iterasi

= {len(t)} titik
""")


# ====================================================================
# BAGIAN 6 — TABEL
# ====================================================================

def tampilkan_tabel(params, t, x, y):
    xp, yp, a = params["xp"], params["yp"], params["a"]
    header = f"{'No':<5}{'t':<10}{'Rumus X':<28}{'Hasil X':<12}{'Rumus Y':<28}{'Hasil Y':<12}"
    garis = "-" * len(header)

    print("  TABEL PERHITUNGAN KOORDINAT PARABOLA\n")
    print("  " + garis + "\n  " + header + "\n  " + garis)

    for i in range(len(t)):
        t_val = t[i]
        rumus_x = f"x={xp}+{a}×({t_val:.2f})²"
        rumus_y = f"y={yp}+2×{a}×({t_val:.2f})"
        print(f"  {i + 1:<5}{t_val:<10.2f}{rumus_x:<28}{x[i]:<12.4f}{rumus_y:<28}{y[i]:<12.4f}")

    print("  " + garis + "\n")

# ====================================================================
# BAGIAN 7 — VISUALISASI MATPLOTLIB
# ====================================================================

def gambar_parabola(params, t_rendah, x_rendah, y_rendah, t_tinggi, x_tinggi, y_tinggi):
    xp, yp, a = params["xp"], params["yp"], params["a"]
    focus_x, focus_y = xp + a, yp

    datasets = [
        {"t": t_rendah, "x": x_rendah, "y": y_rendah, "step": params["step_rendah"], "label": "Resolusi Rendah", "no_gbr": 1},
        {"t": t_tinggi, "x": x_tinggi, "y": y_tinggi, "step": params["step_tinggi"], "label": "Resolusi Tinggi", "no_gbr": 2},
    ]

    for ds in datasets:
        t, x, y = ds["t"], ds["x"], ds["y"]

        fig, ax = plt.subplots(figsize=(14, 8))
        fig.patch.set_facecolor('#0d1117')
        ax.set_facecolor('#0d1117')
        
        fig.suptitle(
            f"KURVA PARABOLA PARAMETRIK (e = 1)\nGambar {ds['no_gbr']} — {ds['label']} (Delta t = {ds['step']})", 
            fontsize=14, color='white', fontweight='bold', y=0.96
        )

        fig.subplots_adjust(left=0.05, right=0.72, top=0.88, bottom=0.08)

        # ── PLOT KURVA DAN TITIK
        ax.plot(x, y, color='yellow', linewidth=2.5, label='Kurva Parabola')
        ax.scatter(x, y, color='white', s=35, zorder=4, label=f'Titik ({len(t)} titik)')

        # ── LABEL TITIK DENGAN 4 TINGKAT ZIG-ZAG ──
        for i in range(len(x)):
            # 4 variasi offset vertikal agar teks menyebar
            offsets_y = [8, -14, 18, -24]
            offset_y = offsets_y[i % 4]
            
            # Ukuran font dikecilkan sedikit jika resolusi tinggi
            ukuran_font = 5.5 if len(x) > 30 else 6.5
            
            ax.annotate(f'({x[i]:.1f},{y[i]:.1f})', (x[i], y[i]), 
                        textcoords='offset points', xytext=(6, offset_y), 
                        fontsize=ukuran_font, color='#aaaaaa')

        # ── VERTEX & FOCUS
        v_offset = (-40, 30) if a > 0 else (40, 30)
        f_offset = (20, -30) if a > 0 else (-20, -30)

        ax.scatter(xp, yp, color='red', s=180, zorder=10)
        ax.annotate(f'VERTEX\n({xp:.2f}, {yp:.2f})', xy=(xp, yp), xytext=v_offset, textcoords='offset points', fontsize=9, color='red', fontweight='bold', arrowprops=dict(arrowstyle='->', color='red', lw=1.5))
        
        ax.scatter(focus_x, focus_y, color='lime', s=130, marker='*', zorder=10)
        ax.annotate(f'FOCUS\n({focus_x:.2f}, {focus_y:.2f})', xy=(focus_x, focus_y), xytext=f_offset, textcoords='offset points', fontsize=9, color='lime', fontweight='bold', arrowprops=dict(arrowstyle='->', color='lime', lw=1.5))

        # ── GARIS BANTU
        ax.axvline(xp, color='cyan', linestyle='--', linewidth=1.5, label=f'Axis Symmetry (x={xp})')
        ax.axhline(yp, color='magenta', linestyle=':', linewidth=1.5, label=f'Garis Vertex (y={yp})')

        # ── SETTING KOTAK PANEL (SIDEBAR KANAN)
        box_style = dict(facecolor='#111827', alpha=0.95, edgecolor='#444444', boxstyle='round,pad=0.6')
        
        panel_rumus = f"PERSAMAAN PARAMETRIK\n{'━'*22}\nx(t) = {xp} + {a}·t²\ny(t) = {yp} + 2·{a}·t"
        fig.text(0.74, 0.88, panel_rumus, fontsize=10, color='cyan', verticalalignment='top', fontfamily='monospace', bbox=box_style)

        arah_bukaan = "Kanan (a > 0)" if a > 0 else "Kiri (a < 0)"
        panel_info = f"INFORMASI KURVA\n{'━'*22}\nVertex      : ({xp}, {yp})\nFocus       : ({focus_x:.2f}, {focus_y:.2f})\nParameter a : {a}\nRentang t   : [{params['t_awal']}, {params['t_akhir']}]\nDelta t     : {ds['step']}\nJml Titik   : {len(t)}\nResolusi    : {ds['label']}\nBukaan      : {arah_bukaan}"
        fig.text(0.74, 0.72, panel_info, fontsize=10, color='yellow', verticalalignment='top', fontfamily='monospace', bbox=box_style)

        t_contoh = t[min(1, len(t) - 1)]
        panel_contoh = f"CONTOH PERHITUNGAN\n{'━'*22}\nt = {t_contoh:.2f}\n\nx = {xp} + {a}×({t_contoh:.2f})²\nx = {xp + a * (t_contoh ** 2):.4f}\n\ny = {yp} + 2×{a}×({t_contoh:.2f})\ny = {yp + 2 * a * t_contoh:.4f}"
        fig.text(0.74, 0.40, panel_contoh, fontsize=10, color='white', verticalalignment='top', fontfamily='monospace', bbox=box_style)

        fig.legend(loc='lower left', bbox_to_anchor=(0.74, 0.08), fontsize=9, facecolor='#111827', edgecolor='#444444', labelcolor='white')

        # ── PENYESUAIAN SUMBU & MARGIN KURVA
        ax.spines['left'].set_position('zero')
        ax.spines['bottom'].set_position('zero')
        ax.spines['right'].set_color('none')
        ax.spines['top'].set_color('none')
        ax.tick_params(colors='#888888')
        ax.grid(True, linestyle='--', alpha=0.2)

        delta_x = abs(x.max() - x.min()) if abs(x.max() - x.min()) > 0 else 10
        delta_y = abs(y.max() - y.min()) if abs(y.max() - y.min()) > 0 else 10
        
        ax.set_xlim(x.min() - delta_x * 0.15, x.max() + delta_x * 0.15)
        ax.set_ylim(y.min() - delta_y * 0.15, y.max() + delta_y * 0.15)
        
        ax.set_xlabel("x", color='white', fontsize=11, loc='right')
        ax.set_ylabel("y", color='white', fontsize=11, loc='top', rotation=0)

    plt.show()

# ====================================================================
# PROGRAM UTAMA
# ====================================================================

def main():
    params = input_parameter()
    validasi_input(params)

    t_rendah = buat_parameter_t(params["t_awal"], params["t_akhir"], params["step_rendah"])
    t_tinggi = buat_parameter_t(params["t_awal"], params["t_akhir"], params["step_tinggi"])

    x_rendah, y_rendah = hitung_koordinat(params, t_rendah)
    x_tinggi, y_tinggi = hitung_koordinat(params, t_tinggi)

    for label, t, x, y in [
        ("RESOLUSI RENDAH", t_rendah, x_rendah, y_rendah),
        ("RESOLUSI TINGGI", t_tinggi, x_tinggi, y_tinggi)
    ]:

        tampilkan_informasi(params, t, label)

        tampilkan_tabel(
            params,
            t,
            x,
            y
        )

    gambar_parabola(params, t_rendah, x_rendah, y_rendah, t_tinggi, x_tinggi, y_tinggi)

    print(f"\n{'='*68}\n  ANALISIS HASIL\n{'='*68}\n")
    print(f"  Resolusi Rendah : Delta t = {params['step_rendah']} → {len(t_rendah)} titik.")
    print(f"  Resolusi Tinggi : Delta t = {params['step_tinggi']} → {len(t_tinggi)} titik.")
    if len(t_rendah) > 0:
        print(f"  Perbandingan    : {len(t_tinggi)} / {len(t_rendah)} = {len(t_tinggi) / len(t_rendah):.1f}×\n")

if __name__ == "__main__":
    main()
```
