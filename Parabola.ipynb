Program Kurva Parabola

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('dark_background')

def plot_parabola():

    print("\n=== GENERATOR KURVA PARABOLA ===\n")

    xp = float(input("Masukkan titik puncak X (xp): "))
    yp = float(input("Masukkan titik puncak Y (yp): "))
    a = float(input("Masukkan nilai a: "))

    step_low = float(input("Step besar (contoh 1.0): "))
    step_high = float(input("Step kecil (contoh 0.25): "))

    fig, axes = plt.subplots(1, 2, figsize=(20, 8))

    fig.suptitle(
        "KURVA PARABOLA (e = 1)\n"
        "Perbandingan Resolusi Berdasarkan Delta t",
        fontsize=16,
        color='white'
    )

    steps = [step_low, step_high]

    titles = [
        "Resolusi Rendah (Step Besar)",
        "Resolusi Tinggi (Step Kecil)"
    ]

    for col in range(2):

        step = steps[col]

        t = np.arange(-10, 10 + step, step)

        x = xp + a * (t**2)
        y = yp + 2 * a * t

        # INFORMASI PARAMETER
        print("\n" + "=" * 100)

        if col == 0:
            print("RESOLUSI RENDAH")
        else:
            print("RESOLUSI TINGGI")

        print("=" * 100)

        print(f"Delta t      = {step}")
        print(f"Jumlah titik = {len(t)}")

        # TABEL PERHITUNGAN
        print("-" * 100)

        print(
            f"{'t':<10}"
            f"{'Perhitungan X':<25}"
            f"{'X':<15}"
            f"{'Perhitungan Y':<25}"
            f"{'Y':<15}"
        )

        print("-" * 100)

        for i in range(len(t)):

            calc_x = f"{xp}+{a}({t[i]:.2f})²"
            calc_y = f"{yp}+2({a})({t[i]:.2f})"

            print(
                f"{t[i]:<10.2f}"
                f"{calc_x:<25}"
                f"{x[i]:<15.2f}"
                f"{calc_y:<25}"
                f"{y[i]:<15.2f}"
            )

        ax = axes[col]

        # KURVA
        ax.plot(
            x,
            y,
            color='yellow',
            linewidth=2,
            marker='o',
            markersize=5,
            label='Kurva Parabola'
        )

        ax.scatter(
            x,
            y,
            color='white',
            s=25,
            zorder=3
        )

        # LABEL SEMUA KOORDINAT
        for i in range(len(x)):

            ax.annotate(
                f'({x[i]:.0f},{y[i]:.0f})',
                (x[i], y[i]),
                textcoords='offset points',
                xytext=(5, 5),
                fontsize=6,
                color='cyan'
            )

        # VERTEX
        ax.scatter(
            xp,
            yp,
            color='red',
            s=150,
            zorder=10
        )

        ax.annotate(
            f'VERTEX\n({xp:.2f},{yp:.2f})',
            xy=(xp, yp),
            xytext=(xp + 2, yp - 2),
            fontsize=10,
            color='red',
            arrowprops=dict(
                arrowstyle='->',
                color='red'
            )
        )

       # AXIS OF SYMMETRY
        ax.axvline(
            xp,
            color='cyan',
            linestyle='--',
            linewidth=2,
            label='Axis of Symmetry'
        )

        ax.axhline(
            yp,
            color='magenta',
            linestyle='--',
            linewidth=1.5,
            label='Garis Vertex'
        )

        # RUMUS
        formula = (
            f"x = {xp} + {a}·t²\n"
            f"y = {yp} + 2·{a}·t"
        )

        ax.text(
            0.02,
            0.95,
            formula,
            transform=ax.transAxes,
            fontsize=10,
            color='cyan',
            verticalalignment='top',
            bbox=dict(
                facecolor='black',
                alpha=0.8
            )
        )

        # INFORMASI RESOLUSI
        info = (
            f"Delta t = {step}\n"
            f"Jumlah titik = {len(t)}"
        )

        ax.text(
            0.02,
            0.80,
            info,
            transform=ax.transAxes,
            fontsize=9,
            color='yellow',
            bbox=dict(
                facecolor='black',
                alpha=0.8
            )
        )

        # CONTOH HITUNG
        t_example = 2

        x_example = xp + a*(t_example**2)
        y_example = yp + 2*a*t_example

        example = (
            f"Contoh Perhitungan\n"
            f"t = {t_example}\n\n"
            f"x = {xp}+{a}({t_example}²)\n"
            f"x = {x_example:.2f}\n\n"
            f"y = {yp}+2({a})({t_example})\n"
            f"y = {y_example:.2f}"
        )

        ax.text(
            0.70,
            0.08,
            example,
            transform=ax.transAxes,
            fontsize=8,
            color='white',
            bbox=dict(
                facecolor='black',
                alpha=0.8
            )
        )

        # SUMBU TENGAH
        ax.spines['left'].set_position('zero')
        ax.spines['bottom'].set_position('zero')

        ax.spines['right'].set_color('none')
        ax.spines['top'].set_color('none')

        ax.tick_params(colors='white')

        ax.grid(
            True,
            linestyle='--',
            alpha=0.3
        )

        # AUTO SCALE
        padding_x = (max(x) - min(x)) * 0.05
        padding_y = (max(y) - min(y)) * 0.05

        ax.set_xlim(
            min(x) - padding_x,
            max(x) + padding_x
        )

        ax.set_ylim(
            min(y) - padding_y,
            max(y) + padding_y
        )

        ax.set_title(
            titles[col],
            fontsize=12,
            color='white'
        )

        ax.legend(loc='upper right')

    plt.tight_layout()
    plt.show()


plot_parabola()
```
