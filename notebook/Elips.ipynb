# 🪐 Program Kurva Elips
> **Parametric Generation of Conic Curves**
> *Visualisasi dan Analisis Pengaruh Nilai $\Delta\theta$ terhadap Resolusi Kurva*

---

## 📌 Deskripsi
Program ini digunakan untuk membangkitkan **kurva elips** menggunakan persamaan parametrik. Selain itu, program ini dirancang untuk membandingkan secara visual dan matematis pengaruh interval **$\Delta\theta$ (delta theta)** terhadap jumlah titik pembentuk dan kualitas *smoothness* (kehalusan) visualisasi kurva.

## 🎯 Tujuan Simulasi
1. Mengimplementasikan persamaan parametrik trigonometri untuk membentuk kurva elips.
2. Menampilkan tabulasi titik-titik koordinat $(x, y)$ pembentuk kurva.
3. Menganalisis perbandingan resolusi grafis berdasarkan variasi nilai $\Delta\theta$.
4. Membuktikan secara komputasional bahwa semakin kecil nilai $\Delta\theta$, maka kerapatan titik semakin tinggi sehingga kurva terbentuk lebih halus.

---

## 📐 Dasar Teori
Elips merupakan salah satu kurva irisan kerucut (*Conic Section*) tertutup yang memiliki nilai eksentrisitas antara 0 dan 1 ($0 \le e < 1$). Pada program komputasi ini, elips dibangkitkan menggunakan **persamaan parametrik** dengan memanfaatkan fungsi trigonometri dasar:

$$x = xc + a \cdot \cos(\theta)$$
$$y = yc + b \cdot \sin(\theta)$$

**Tabel Parameter:**

| Simbol | Parameter | Keterangan |
| :---: | :--- | :--- |
| **$xc, yc$** | Titik Pusat | Koordinat pusat elips pada sumbu Kartesian |
| **$a$** | Jari-jari Horizontal | Jarak dari pusat ke batas ujung sumbu $X$ |
| **$b$** | Jari-jari Vertikal | Jarak dari pusat ke batas ujung sumbu $Y$ |
| **$\theta$** | Sudut Parameter | Parameter sudut pembentuk kurva (dalam radian) |
| **$\Delta\theta$** | *Step Size* | Interval perubahan parameter sudut $\theta$ |

---

## 🔍 Analisis Pengaruh $\Delta\theta$ (Resolusi)
Parameter $\theta$ dibangkitkan dari rentang interval **$0$ sampai $2\pi$** (membentuk 360 derajat atau satu putaran penuh tertutup). Nilai $\Delta\theta$ bertindak sebagai *step size* yang menentukan jarak antar titik komputasi.

* **Resolusi Rendah ($\Delta\theta$ Besar):**
  Contoh jika $\Delta\theta = 1.0$ radian, program hanya akan meng-*generate* sedikit titik (sekitar 7 titik). Akibatnya, grafik akan terlihat patah-patah seperti poligon kasar.
* **Resolusi Tinggi ($\Delta\theta$ Kecil):**
  Contoh jika $\Delta\theta = 0.1$ radian, program akan menghasilkan banyak titik (sekitar 63 titik rapat). Grafik elips akan terlihat melengkung sempurna dan halus (*smooth*).

---

## ⚙️ Algoritma Program
1. **Inisialisasi Input:** Membaca nilai pusat $(xc, yc)$ serta jari-jari mendatar dan tegak $(a, b)$.
2. **Konfigurasi Resolusi:** Menentukan nilai $\Delta\theta$ besar (untuk *drafting*) dan $\Delta\theta$ kecil (untuk rendering presisi).
3. **Pembangkitan Sudut:** Membuat deret *array* parameter $\theta$ dari $0$ hingga $2\pi$.
4. **Kalkulasi Koordinat:** Menghitung matriks titik $x$ dan $y$ secara simultan menggunakan persamaan parametrik trigonometri.
5. **Tabulasi Data:** Menampilkan hasil iterasi perhitungan titik ke dalam bentuk tabel.
6. **Visualisasi (Plotting):** Menggambar kurva menggunakan pustaka grafis dengan membandingkan dua resolusi secara berdampingan (*side-by-side*).

---

## 💡 Kesimpulan
Pendekatan **parametrik trigonometri** sangat efektif untuk membangkitkan kurva tertutup seperti elips tanpa mengalami masalah fungsi ganda pada sumbu $Y$. Resolusi visual dari komputasi grafik ini sangat bergantung pada penentuan parameter **$\Delta\theta$**. Optimalisasi nilai *step size* ini penting untuk menyeimbangkan antara kualitas visual (kehalusan kurva) dan beban komputasi (jumlah iterasi memori).

```python
import numpy as np
import matplotlib.pyplot as plt

# Mengatur tema gelap bawaan matplotlib
plt.style.use('dark_background')
plt.rcParams['toolbar'] = 'toolbar2'

def plot_elips():
    # --- HEADER TERMINAL ---
    print("================================================================================")
    print(" GENERATOR KURVA ELIPS PARAMETRIK")
    print(" Grafika Komputer - Semester 4")
    print("================================================================================\n")

    # Input User
    xc = float(input("Masukkan titik pusat X (xc): "))
    yc = float(input("Masukkan titik pusat Y (yc): "))
    a  = float(input("Masukkan jari-jari horizontal (a): "))
    b  = float(input("Masukkan jari-jari vertikal (b): "))

    n_low  = int(input("Jumlah titik resolusi rendah  (contoh 11): "))
    n_high = int(input("Jumlah titik resolusi tinggi  (contoh 21): "))

    configs = [
        ("Resolusi Rendah", n_low),
        ("Resolusi Tinggi", n_high),
    ]

    # Menghitung jarak fokus (c)
    c = np.sqrt(abs(a**2 - b**2))

    for title, n_pts in configs:
        # Array sudut theta dari 0 sampai 2π
        theta = np.linspace(0, 2 * np.pi, n_pts, endpoint=True)
        step  = 2 * np.pi / (n_pts - 1)

        # Persamaan Parametrik Elips
        x = xc + a * np.cos(theta)
        y = yc + b * np.sin(theta)

        # ── 1. OUTPUT TERMINAL ──
        print("\n" + "=" * 80)
        print(f" INFORMASI PARAMETER ELIPS - {title.upper()}")
        print("=" * 80 + "\n")
        
        print(f"CENTER = ({xc}, {yc})\n")
        
        print("Parameter Jari-jari")
        print("-------------------")
        print(f"Horizontal (a) = {a}")
        print(f"Vertikal   (b) = {b}")
        print(f"Jarak Fokus(c) = {c:.4f}\n")

        print("Persamaan Parametrik")
        print("--------------------")
        print(f"x(θ) = {xc} + {a} * cos(θ)")
        print(f"y(θ) = {yc} + {b} * sin(θ)\n")

        print("-" * 115)
        print(f"{'No':<5}{'θ (rad)':<10}{'Perhitungan X':<30}{'X':<15}{'Perhitungan Y':<35}{'Y':<15}")
        print("-" * 115)
        for i in range(len(theta)):
            calc_x = f"x={xc}+{a}*cos({theta[i]:.2f})"
            calc_y = f"y={yc}+{b}*sin({theta[i]:.2f})"
            print(f"{i+1:<5}{theta[i]:<10.2f}{calc_x:<30}{x[i]:<15.4f}{calc_y:<35}{y[i]:<15.4f}")
        print("-" * 115)

        # ── 2. PLOT MATPLOTLIB ──
        fig, ax = plt.subplots(figsize=(14, 7.5))
        fig.subplots_adjust(left=0.08, right=0.68, top=0.90, bottom=0.12)

        # Judul Gambar
        fig.suptitle(
            f"KURVA ELIPS PARAMETRIK\nGambar - {title} (Delta θ = {step:.2f})",
            fontsize=13, fontweight='bold', color='white', y=0.97
        )

        # Plot Kurva Utama
        ax.plot(x, y, color='#FFFF00', linewidth=2, zorder=2, label='Kurva Elips')
        ax.scatter(x, y, color='white', s=30, zorder=3, label=f'Titik ({n_pts} titik)')

        # Sumbu Pembantu
        ax.axvline(xc, color='#00FFFF', linestyle='--', linewidth=1.2, label=f'Sumbu Vertikal (x={xc})')
        ax.axhline(yc, color='#FF00FF', linestyle='--', linewidth=1.2, label=f'Sumbu Horizontal (y={yc})')

        # Titik Pusat (Center)
        ax.scatter(xc, yc, color='red', s=100, zorder=5)
        ax.text(xc - (a*0.05), yc + (b*0.05), f"CENTER\n({xc:.2f}, {yc:.2f})", 
                color='red', fontsize=9, fontweight='bold', ha='right')

        # --- PENANDA TITIK AWAL & AKHIR (DIREVISI) ---
        # Menambahkan koordinat (x, y) langsung pada label penanda awal dan akhir
        ax.scatter(x[0], y[0], color='#00FF00', marker='s', s=80, zorder=6, label='Titik Awal (θ=0)')
        ax.annotate(f"TITIK AWAL\n({x[0]:.2f}, {y[0]:.2f})", xy=(x[0], y[0]), xytext=(x[0] + a*0.15, y[0] + b*0.25),
                    color='#00FF00', fontsize=9, fontweight='bold', ha='left',
                    arrowprops=dict(arrowstyle='->', color='#00FF00', lw=1.5))

        ax.scatter(x[-1], y[-1], color='#FF9900', marker='x', s=80, zorder=7, label='Titik Akhir (θ=2π)')
        ax.annotate(f"TITIK AKHIR\n({x[-1]:.2f}, {y[-1]:.2f})", xy=(x[-1], y[-1]), xytext=(x[-1] + a*0.15, y[-1] - b*0.25),
                    color='#FF9900', fontsize=9, fontweight='bold', ha='left',
                    arrowprops=dict(arrowstyle='->', color='#FF9900', lw=1.5))

        # --- KETERANGAN KOORDINAT TIAP TITIK ---
        # Looping dimulai dari indeks 1 sampai n-2 saja agar tidak bentrok dengan teks TITIK AWAL/AKHIR
        for i in range(1, len(theta) - 1):
            ax.annotate(f"({x[i]:.1f}, {y[i]:.1f})",
                        (x[i], y[i]),
                        textcoords="offset points",
                        xytext=(6, 6),
                        fontsize=7,
                        color='#A0A0A0',
                        ha='left')

        # Format kotak teks UI
        bbox_props = dict(boxstyle="round,pad=0.6", fc="#0B1320", ec="#00FFFF", lw=1, alpha=0.9)

        # Panel Kanan 1: PERSAMAAN PARAMETRIK
        eq_text = (
            "PERSAMAAN PARAMETRIK\n"
            "--------------------\n"
            f"x(θ) = {xc} + {a}·cos(θ)\n"
            f"y(θ) = {yc} + {b}·sin(θ)"
        )
        fig.text(0.72, 0.82, eq_text, fontsize=9, color='#00FFFF', 
                 family='monospace', bbox=bbox_props, va='top')

        # Panel Kanan 2: INFORMASI KURVA
        info_text = (
            "INFORMASI KURVA\n"
            "--------------------\n"
            f"Center       : ({xc:.2f}, {yc:.2f})\n"
            f"Radius a (x) : {a}\n"
            f"Radius b (y) : {b}\n"
            f"Rentang θ    : [0.0, 2π]\n"
            f"Delta θ      : {step:.4f}\n"
            f"Jml Titik    : {n_pts}\n"
            f"Resolusi     : {title}\n"
        )
        fig.text(0.72, 0.65, info_text, fontsize=9, color='#FFFF00', 
                 family='monospace', bbox=bbox_props, va='top')

        # Panel Kanan 3: CONTOH PERHITUNGAN
        idx = n_pts // 2
        te, xe, ye = theta[idx], x[idx], y[idx]
        calc_text = (
            "CONTOH PERHITUNGAN\n"
            "--------------------\n"
            f"θ = {te:.2f} rad\n\n"
            f"x = {xc} + {a}·cos({te:.2f})\n"
            f"x = {xe:.4f}\n\n"
            f"y = {yc} + {b}·sin({te:.2f})\n"
            f"y = {ye:.4f}"
        )
        fig.text(0.72, 0.35, calc_text, fontsize=9, color='white', 
                 family='monospace', bbox=bbox_props, va='top')

        # Setting area grafik
        ax.set_aspect('equal', adjustable='box')
        ax.set_xlim(xc - a * 1.5, xc + a * 1.5)
        ax.set_ylim(yc - b * 1.5, yc + b * 1.5)

        # Pengaturan Grid
        ax.grid(True, linestyle=':', alpha=0.3, color='white')
        ax.spines['left'].set_position('zero')
        ax.spines['bottom'].set_position('zero')
        ax.spines['right'].set_color('none')
        ax.spines['top'].set_color('none')
        ax.tick_params(colors='white')

        # Legend
        ax.legend(loc='lower right', facecolor='#0B1320', edgecolor='#444444', labelcolor='white', fontsize=8)

        plt.show() 
    #
    # ── 3. ANALISIS HASIL DI TERMINAL ──
    print("\n" + "=" * 80)
    print(" ANALISIS HASIL")
    print("=" * 80 + "\n")
    print(f" Resolusi Rendah : Delta θ = {2 * np.pi / (n_low - 1):.4f} rad -> {n_low} titik.")
    print(f" Resolusi Tinggi : Delta θ = {2 * np.pi / (n_high - 1):.4f} rad -> {n_high} titik.")
    print(f" Perbandingan    : {n_high} / {n_low} = {n_high/n_low:.1f}x\n")

if __name__ == "__main__":
    plot_elips()
```
