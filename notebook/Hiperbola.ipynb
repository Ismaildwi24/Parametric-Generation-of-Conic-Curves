# Program Kurva Hiperbola

# Parametric Generation of Conic Curves

# Deskripsi
Program ini digunakan untuk membangkitkan kurva hiperbola menggunakan persamaan parametrik serta membandingkan pengaruh nilai Δθ (delta theta) terhadap jumlah titik dan kualitas visualisasi kurva.

# Tujuan
1. Membuat kurva hiperbola menggunakan persamaan parametrik (trigonometri secant dan tangent).
2. Menampilkan titik-titik pembentuk kedua cabang kurva (kiri dan kanan).
3. Membandingkan resolusi kurva berdasarkan nilai Δθ.
4. Menunjukkan bahwa semakin kecil Δθ maka semakin banyak titik yang dihasilkan sehingga kurva menjadi lebih halus (smooth).

# Dasar Teori
Hiperbola merupakan kurva irisan kerucut (Conic Section) yang memiliki eksentrisitas e > 1 dan terdiri dari dua cabang yang terpisah. Pada program ini hiperbola dibangkitkan menggunakan persamaan parametrik:

x = xc ± a · sec(θ)  
y = yc + b · tan(θ)

Keterangan:
- xc = koordinat x titik pusat (center)
- yc = koordinat y titik pusat (center)
- a = semi-sumbu transversal (menentukan jarak titik pusat ke puncak kurva)
- b = semi-sumbu konjugasi (menentukan bukaan kurva)
- θ = parameter sudut pembentuk kurva (dalam radian)
- Δθ = interval perubahan parameter θ

# Pengaruh Delta Theta (Δθ) terhadap Resolusi
Parameter θ dibangkitkan dari interval mendekati -π/2 sampai π/2. Nilai Δθ menentukan jarak perubahan parameter sudut θ.

Contoh:
- Δθ = 15° menghasilkan jumlah titik yang lebih sedikit (resolusi rendah/patah-patah).
- Δθ = 1° menghasilkan jumlah titik yang jauh lebih banyak (resolusi tinggi/halus).

Semakin kecil nilai Δθ maka jumlah titik yang dihasilkan semakin banyak sehingga kedua cabang kurva hiperbola menjadi lebih halus (smooth).

# Algoritma Program
1. Menginput nilai xc, yc, a, dan b.
2. Menginput nilai Δθ besar dan Δθ kecil.
3. Membentuk parameter θ dari rentang -90° hingga 90°.
4. Menghitung koordinat x dan y untuk cabang kanan dan cabang kiri menggunakan persamaan parametrik trigonometri.
5. Menampilkan tabel perhitungan titik-titik kurva.
6. Menggambar kurva hiperbola beserta titik-titiknya.
7. Membandingkan hasil resolusi rendah dan resolusi tinggi.

# Kesimpulan
Kurva hiperbola dapat dibangkitkan menggunakan persamaan parametrik dengan memanfaatkan parameter sudut θ. Nilai Δθ memengaruhi jumlah titik yang dihasilkan. Semakin kecil nilai Δθ maka semakin banyak titik yang terbentuk sehingga visualisasi kedua cabang kurva menjadi lebih halus dan presisi.

```python
from matplotlib import style
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('dark_background')

def plot_hyperbola():
    print("\n=== GENERATOR KURVA HIPERBOLA ===\n")

    xc = float(input("Masukkan titik pusat X (xc): "))
    yc = float(input("Masukkan titik pusat Y (yc): "))
    a = float(input("Masukkan semi-sumbu transversal (a): "))
    b = float(input("Masukkan semi-sumbu konjugasi (b): "))

    step_low = float(input("Step besar (dalam derajat, contoh 15): "))
    step_high = float(input("Step kecil (dalam derajat, contoh 1): "))

    steps = [step_low, step_high]

    titles = [
        "Resolusi Rendah (Step Besar)",
        "Resolusi Tinggi (Step Kecil)"
    ]

    for col in range(2):
        step_deg = steps[col]

        fig, ax = plt.subplots(figsize=(10, 8))
        fig.suptitle(
            "KURVA HIPERBOLA (e > 1)\n"
            "Perbandingan Resolusi Berdasarkan Delta Theta (θ)",
            fontsize=14,
            color='white'
        )
        
        start_deg = -89.9
        end_deg = 89.9
        theta_deg = np.arange(start_deg, end_deg + (step_deg*0.1), step_deg)
        theta_rad = np.deg2rad(theta_deg)

        x_right = xc + a * (1 / np.cos(theta_rad))
        y_right = yc + b * np.tan(theta_rad)
        
        x_left = xc - a * (1 / np.cos(theta_rad))
        y_left = yc + b * np.tan(theta_rad)

        print("\n" + "=" * 100)

        if col == 0:
            print("RESOLUSI RENDAH")
        else:
            print("RESOLUSI TINGGI")

        print("=" * 100)

        # Proses perhitungan jumlah titik
        range_deg = end_deg - start_deg
        num_points = len(theta_deg)
        print(f"Delta θ      = {step_deg}°")
        print(f"Proses cari jumlah titik: Rentang = {end_deg}° - ({start_deg}°) = {range_deg:.1f}°")
        print(f"Jumlah titik = {range_deg:.1f}° / {step_deg}° ≈ {num_points} (per cabang)")

        # Perhitungan Titik Fokus
        c = np.sqrt(a**2 + b**2)
        print("-" * 100)
        print("PERHITUNGAN TITIK FOKUS")
        print(f"c = √(a² + b²) = √({a}² + {b}²) = √({a**2} + {b**2}) = {c:.2f}")
        print(f"Fokus Kanan (F1) = (xc + c, yc) = ({xc} + {c:.2f}, {yc}) = ({xc + c:.2f}, {yc})")
        print(f"Fokus Kiri (F2)  = (xc - c, yc) = ({xc} - {c:.2f}, {yc}) = ({xc - c:.2f}, {yc})")

        # Perhitungan Asimtot
        print("-" * 100)
        print("PERHITUNGAN ASIMTOT")
        print(f"Persamaan Asimtot 1: y = (b/a)(x - xc) + yc = ({b}/{a})(x - {xc}) + {yc}")
        print(f"Persamaan Asimtot 2: y = -(b/a)(x - xc) + yc = -({b}/{a})(x - {xc}) + {yc}")

        print("-" * 100)
        print(
            f"{'θ (deg)':<10}"
            f"{'Perhitungan X (Kanan)':<30}"
            f"{'X':<15}"
            f"{'Perhitungan Y':<25}"
            f"{'Y':<15}"
        )
        print("-" * 100)

        for i in range(len(theta_deg)):
            calc_x = f"{xc}+{a}*sec({theta_deg[i]:.1f}°)"
            calc_y = f"{yc}+{b}*tan({theta_deg[i]:.1f}°)"

            print(
                f"{theta_deg[i]:<10.1f}"
                f"{calc_x:<30}"
                f"{x_right[i]:<15.2f}"
                f"{calc_y:<25}"
                f"{y_right[i]:<15.2f}"
            )

        ax.plot(x_right, y_right, color='yellow', linewidth=2, marker='o', markersize=5, label='Cabang Kanan')
        ax.scatter(x_right, y_right, color='white', s=25, zorder=3)

        ax.plot(x_left, y_left, color='#00ffcc', linewidth=2, marker='o', markersize=5, label='Cabang Kiri')
        ax.scatter(x_left, y_left, color='white', s=25, zorder=3)

        # Plot Titik Fokus
        f1_x, f1_y = xc + c, yc
        f2_x, f2_y = xc - c, yc
        ax.scatter([f1_x, f2_x], [f1_y, f2_y], color='orange', s=100, zorder=5, marker='X', label='Titik Fokus')
        ax.annotate(f'F1({f1_x:.1f}, {f1_y:.1f})', (f1_x, f1_y), textcoords='offset points', xytext=(5, 5), fontsize=8, color='orange')
        ax.annotate(f'F2({f2_x:.1f}, {f2_y:.1f})', (f2_x, f2_y), textcoords='offset points', xytext=(-35, 5), fontsize=8, color='orange')

        # Plot Asimtot
        x_asymptote = np.array([xc - 15*a, xc + 15*a])
        y_asymp_1 = (b/a) * (x_asymptote - xc) + yc
        y_asymp_2 = -(b/a) * (x_asymptote - xc) + yc
        ax.plot(x_asymptote, y_asymp_1, color='lightgreen', linestyle=':', linewidth=2, label='Asimtot')
        ax.plot(x_asymptote, y_asymp_2, color='lightgreen', linestyle=':', linewidth=2)

        limit_x_label = xc + 15 * a
        limit_y_label = yc + 15 * b
        
        for i in range(len(theta_deg)):
            if abs(x_right[i]) <= limit_x_label and abs(y_right[i]) <= limit_y_label:
                ax.annotate(f'({x_right[i]:.0f},{y_right[i]:.0f})', (x_right[i], y_right[i]),
                            textcoords='offset points', xytext=(5, 5), fontsize=6, color='cyan')
            if abs(x_left[i]) <= limit_x_label and abs(y_left[i]) <= limit_y_label:
                ax.annotate(f'({x_left[i]:.0f},{y_left[i]:.0f})', (x_left[i], y_left[i]),
                            textcoords='offset points', xytext=(5, 5), fontsize=6, color='cyan')

        ax.scatter(xc, yc, color='red', s=150, zorder=10)
        ax.annotate(f'CENTER\n({xc:.2f},{yc:.2f})', xy=(xc, yc), xytext=(xc + a*0.5, yc - b*1.5),
                    fontsize=10, color='red', arrowprops=dict(arrowstyle='->', color='red'))

        ax.axvline(xc, color='cyan', linestyle='--', linewidth=2, label='Sumbu Konjugasi Y')
        ax.axhline(yc, color='magenta', linestyle='--', linewidth=1.5, label='Sumbu Transversal X')

        formula = (
            f"x = {xc} ± {a}·sec(θ)\n"
            f"y = {yc} + {b}·tan(θ)\n"
            f"Asimtot:\ny = ±({b}/{a})(x - {xc}) + {yc}"
        )

        ax.text(0.02, 0.98, formula, transform=ax.transAxes, fontsize=10, color='cyan',
                verticalalignment='top', bbox=dict(facecolor='black', alpha=0.8))

        info = (
            f"Delta θ = {step_deg}°\n"
            f"Jml titik/cabang = {num_points}"
        )

        ax.text(0.02, 0.82, info, transform=ax.transAxes, fontsize=9, color='yellow',
                bbox=dict(facecolor='black', alpha=0.8))

        ax.spines['left'].set_position('zero')
        ax.spines['bottom'].set_position('zero')
        ax.spines['right'].set_color('none')
        ax.spines['top'].set_color('none')
        ax.tick_params(colors='white')
        ax.grid(True, linestyle='--', alpha=0.3)

        ax.set_xlim(xc - 5*a, xc + 5*a)
        ax.set_ylim(yc - 5*b, yc + 5*b)

        ax.set_title(titles[col], fontsize=12, color='white')
        ax.legend(loc='upper right')

        plt.tight_layout()

    plt.show()

plot_hyperbola()
```
