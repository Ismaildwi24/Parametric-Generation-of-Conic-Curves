# Program Kurva Hiperbola

# Parametric Generation of Conic Curves

# Deskripsi
Program ini digunakan untuk membangkitkan kurva hiperbola menggunakan persamaan parametrik serta membandingkan pengaruh nilai Δθ (delta theta) terhadap jumlah titik dan kualitas visualisasi kurva.

# Tujuan
1. Membuat kurva hiperbola menggunakan persamaan parametrik (trigonometri secant dan tangent).
2. Menampilkan titik-titik pembentuk kedua cabang kurva (kiri dan kanan).
3. Membandingkan resolusi kurva berdasarkan nilai Δθ.
4. Menunjukkan bahwa semakin kecil Δθ maka semakin banyak titik yang dihasilkan sehingga kurva menjadi lebih halus (smooth).

# Dasar Teori
Hiperbola merupakan kurva irisan kerucut (Conic Section) yang memiliki eksentrisitas e > 1 dan terdiri dari dua cabang yang terpisah. Pada program ini hiperbola dibangkitkan menggunakan persamaan parametrik:

x = xc ± a · sec(θ)  
y = yc + b · tan(θ)

Keterangan:
- xc = koordinat x titik pusat (center)
- yc = koordinat y titik pusat (center)
- a = semi-sumbu transversal (menentukan jarak titik pusat ke puncak kurva)
- b = semi-sumbu konjugasi (menentukan bukaan kurva)
- θ = parameter sudut pembentuk kurva (dalam radian)
- Δθ = interval perubahan parameter θ

# Pengaruh Delta Theta (Δθ) terhadap Resolusi
Parameter θ dibangkitkan dari interval mendekati -π/2 sampai π/2. Nilai Δθ menentukan jarak perubahan parameter sudut θ.

Contoh:
- Δθ = 0.3 menghasilkan jumlah titik yang lebih sedikit (resolusi rendah/patah-patah).
- Δθ = 0.05 menghasilkan jumlah titik yang jauh lebih banyak (resolusi tinggi/halus).

Semakin kecil nilai Δθ maka jumlah titik yang dihasilkan semakin banyak sehingga kedua cabang kurva hiperbola menjadi lebih halus (smooth).

# Algoritma Program
1. Menginput nilai xc, yc, a, dan b.
2. Menginput nilai Δθ besar dan Δθ kecil.
3. Membentuk parameter θ dari rentang -π/2 hingga π/2 (dengan sedikit batas margin/epsilon untuk menghindari pembagian dengan nol).
4. Menghitung koordinat x dan y untuk cabang kanan dan cabang kiri menggunakan persamaan parametrik trigonometri.
5. Menampilkan tabel perhitungan titik-titik kurva.
6. Menggambar kurva hiperbola beserta titik-titiknya.
7. Membandingkan hasil resolusi rendah dan resolusi tinggi.

# Kesimpulan
Kurva hiperbola dapat dibangkitkan menggunakan persamaan parametrik dengan memanfaatkan parameter sudut θ. Nilai Δθ memengaruhi jumlah titik yang dihasilkan. Semakin kecil nilai Δθ maka semakin banyak titik yang terbentuk sehingga visualisasi kedua cabang kurva menjadi lebih halus dan presisi.

```python
from matplotlib import style
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('dark_background')

def plot_hyperbola():

    print("\n=== GENERATOR KURVA HIPERBOLA ===\n")

    xc = float(input("Masukkan titik pusat X (xc): "))
    yc = float(input("Masukkan titik pusat Y (yc): "))
    a = float(input("Masukkan semi-sumbu transversal (a): "))
    b = float(input("Masukkan semi-sumbu konjugasi (b): "))

    step_low = float(input("Step besar (contoh 0.3): "))
    step_high = float(input("Step kecil (contoh 0.05): "))

    fig, axes = plt.subplots(1, 2, figsize=(20, 8))

    fig.suptitle(
        "KURVA HIPERBOLA (e > 1)\n"
        "Perbandingan Resolusi Berdasarkan Delta Theta (θ)",
        fontsize=16,
        color='white'
    )

    steps = [step_low, step_high]

    titles = [
        "Resolusi Rendah (Step Besar)",
        "Resolusi Tinggi (Step Kecil)"
    ]

    for col in range(2):
        step = steps[col]
        
        epsilon = 0.01
        theta = np.arange(-np.pi/2 + epsilon, np.pi/2 - epsilon, step)

        x_right = xc + a * (1 / np.cos(theta))
        y_right = yc + b * np.tan(theta)
        
        x_left = xc - a * (1 / np.cos(theta))
        y_left = yc + b * np.tan(theta)

        print("\n" + "=" * 100)

        if col == 0:
            print("RESOLUSI RENDAH")
        else:
            print("RESOLUSI TINGGI")

        print("=" * 100)

        print(f"Delta θ      = {step}")
        print(f"Jumlah titik = {len(theta)} (per cabang)")

        print("-" * 100)

        print(
            f"{'θ (rad)':<10}"
            f"{'Perhitungan X (Kanan)':<30}"
            f"{'X':<15}"
            f"{'Perhitungan Y':<25}"
            f"{'Y':<15}"
        )

        print("-" * 100)

        for i in range(len(theta)):
            calc_x = f"{xc}+{a}*sec({theta[i]:.2f})"
            calc_y = f"{yc}+{b}*tan({theta[i]:.2f})"

            print(
                f"{theta[i]:<10.2f}"
                f"{calc_x:<30}"
                f"{x_right[i]:<15.2f}"
                f"{calc_y:<25}"
                f"{y_right[i]:<15.2f}"
            )

        ax = axes[col]

        ax.plot(
            x_right,
            y_right,
            color='yellow',
            linewidth=2,
            marker='o',
            markersize=5,
            label='Cabang Kanan'
        )
        ax.scatter(x_right, y_right, color='white', s=25, zorder=3)

        ax.plot(
            x_left,
            y_left,
            color='#00ffcc',
            linewidth=2,
            marker='o',
            markersize=5,
            label='Cabang Kiri'
        )
        ax.scatter(x_left, y_left, color='white', s=25, zorder=3)

        limit_x_label = xc + 15 * a
        limit_y_label = yc + 15 * b
        
        for i in range(len(theta)):
            if abs(x_right[i]) <= limit_x_label and abs(y_right[i]) <= limit_y_label:
                ax.annotate(f'({x_right[i]:.0f},{y_right[i]:.0f})', (x_right[i], y_right[i]),
                            textcoords='offset points', xytext=(5, 5), fontsize=6, color='cyan')
            if abs(x_left[i]) <= limit_x_label and abs(y_left[i]) <= limit_y_label:
                ax.annotate(f'({x_left[i]:.0f},{y_left[i]:.0f})', (x_left[i], y_left[i]),
                            textcoords='offset points', xytext=(5, 5), fontsize=6, color='cyan')

        ax.scatter(
            xc,
            yc,
            color='red',
            s=150,
            zorder=10
        )

        ax.annotate(
            f'CENTER\n({xc:.2f},{yc:.2f})',
            xy=(xc, yc),
            xytext=(xc + a*0.5, yc - b*1.5),
            fontsize=10,
            color='red',
            arrowprops=dict(
                arrowstyle='->',
                color='red'
            )
        )

        ax.axvline(
            xc,
            color='cyan',
            linestyle='--',
            linewidth=2,
            label='Sumbu Konjugasi Y'
        )

        ax.axhline(
            yc,
            color='magenta',
            linestyle='--',
            linewidth=1.5,
            label='Sumbu Transversal X'
        )

        formula = (
            f"x = {xc} ± {a}·sec(θ)\n"
            f"y = {yc} + {b}·tan(θ)"
        )

        ax.text(
            0.02,
            0.95,
            formula,
            transform=ax.transAxes,
            fontsize=10,
            color='cyan',
            verticalalignment='top',
            bbox=dict(
                facecolor='black',
                alpha=0.8
            )
        )

        info = (
            f"Delta θ = {step}\n"
            f"Jumlah titik/cabang = {len(theta)}"
        )

        ax.text(
            0.02,
            0.80,
            info,
            transform=ax.transAxes,
            fontsize=9,
            color='yellow',
            bbox=dict(
                facecolor='black',
                alpha=0.8
            )
        )

        theta_example = theta[len(theta)//2]
        x_example = xc + a * (1/np.cos(theta_example))
        y_example = yc + b * np.tan(theta_example)

        example = (
            f"Contoh Perhitungan Kanan\n"
            f"θ = {theta_example:.2f} rad\n\n"
            f"x = {xc}+{a}·sec({theta_example:.2f})\n"
            f"x = {x_example:.2f}\n\n"
            f"y = {yc}+{b}·tan({theta_example:.2f})\n"
            f"y = {y_example:.2f}"
        )

        ax.text(
            0.75,
            0.05,
            example,
            transform=ax.transAxes,
            fontsize=8,
            color='white',
            bbox=dict(
                facecolor='black',
                alpha=0.8
            )
        )

        ax.spines['left'].set_position('zero')
        ax.spines['bottom'].set_position('zero')

        ax.spines['right'].set_color('none')
        ax.spines['top'].set_color('none')

        ax.tick_params(colors='white')

        ax.grid(
            True,
            linestyle='--',
            alpha=0.3
        )

        ax.set_xlim(xc - 10*a, xc + 10*a)
        ax.set_ylim(yc - 10*b, yc + 10*b)

        ax.set_title(
            titles[col],
            fontsize=12,
            color='white'
        )

        ax.legend(loc='upper right')

    plt.tight_layout()
    plt.show()

plot_hyperbola()
```
